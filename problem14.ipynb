```python
def Collatz(N):
    if N % 2 ==0:
        return N//2
    else:
        return N*3+1
    
def CollatzSize(N):
    if N ==0: return 0
    
    count =2 # the first and last are technichly part of this
    # for example 1 -> 1 has 2 members in it but while is never active
    while N != 1:
        N = Collatz(N)
        count+=1
    return count

CollatzSize(1)
```

### I feel like graphing this

```python
import matplotlib.pyplot as plt

GraphingData = list(enumerate(map(CollatzSize,range(10**3))))
x =list( map(lambda p:p[0],GraphingData))
y =list( map(lambda p:p[1],GraphingData))

plt.scatter(x,y,s=2)

plt.show()
```
