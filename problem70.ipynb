Euler's totient function, $\phi(n)$ [sometimes called the phi function], is used to determine the number of positive numbers less than or equal to $n$ which are relatively prime to $n$. For example, as $1, 2, 4, 5, 7$, and $8$, are all less than nine and relatively prime to nine, $\phi(9)=6$.<br>The number $1$ is considered to be relatively prime to every positive number, so $\phi(1)=1$. 
Interestingly, $\phi(87109)=79180$, and it can be seen that $87109$ is a permutation of $79180$.
Find the value of $n$, $1 \lt n \lt 10^7$, for which $\phi(n)$ is a permutation of $n$ and the ratio $n/\phi(n)$ produces a minimum.


```python
def hcf(x, y):
    while y:
        x, y = y, x % y

    return x

def isCoprime(x,y):
    return hcf(x,y) == 1

def phi(x):
    return sum(map(lambda i:isCoprime(x,i) ,range(1,x)))

def isPermute(x,y):
    x = str(x)
    y = str(y)
    if len(x) != len(y):return False
    return sorted(x) == sorted(y)
    
    
    
        
```

```python
import matplotlib.pyplot as plt

plt.plot(list(map(phi,range(10**3))),'bo')
#plt.plot(list(map(lambda x: x/1,range(10**3))),color = "white")
#plt.plot(list(map(lambda x: x/2,range(10**3)))  ,color = "white")
#plt.plot(list(map(lambda x: x/3,range(10**3)))  ,color = "white")
#plt.plot(list(map(lambda x: x*2/3,range(10**3))),color = "white")

plt.ylabel("phi(x)")
plt.xlabel("x")
```

```python
N = range(2,10**7)

Coordinates = map(lambda N:(N,phi(N)),N)

Coordinates = filter(lambda x: isPermute(x[0],x[1]),Coordinates)



print(min(Coordinates,key=lambda x: x[0]/x[1]))

```
